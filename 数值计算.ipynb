```python
import cv2

img_cat = cv2.imread("./data/images/cat.png")
img_dog = cv2.imread("./data/images/dog.png")

img_cat2 = img_cat + 100
```

```python
# 超出255相当于 %256
(img_cat + img_cat2)[:5,:,0]
```

```python
# 重定义图片分辨率
img_cat_resize = cv2.resize(img_cat, (1280, 800))
img_cat_resize.shape
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_cat_resize)
```

```python
img_cat_resize2 = cv2.resize(img_cat, (0, 0), fx=1, fy=3)
plt.imshow(img_cat_resize2)
```

```python
# 图像融合 R = ax1 + bx2 + b a,b是权重
res = cv2.addWeighted(img_cat_resize, 0.4, img_dog, 0.6, 0)
plt.imshow(res)
```
