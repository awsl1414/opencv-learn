```python
import cv2
from utils import cv_show
```

```python
img_role = cv2.imread("./data/images/role.png")
```

```python
cv_show("role", img_role)
img_role.shape
```

```python
# ()
role_up = cv2.pyrUp(img_role)
cv_show("role_up", role_up)
role_up.shape
```

```python
# ()
role_down = cv2.pyrDown(img_role)
cv_show("role_down", role_down)
role_down.shape
```

```python
role_down_up = cv2.pyrUp(role_down)

# 
role_laplacian = img_role - role_down_up
cv_show("role_laplacian", role_laplacian)
```
